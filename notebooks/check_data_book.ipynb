```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/quran/Desktop/Data_Sci/ServerLifecycleML/data/synthetic_server_data.csv")
print(df.head())
```

```python
df.info()  # Check data types and missing values
df.describe()  # Summary statistics
df["OS"].value_counts()  # Count servers by OS
df["Data_Center"].value_counts()  # Count servers by data center
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Failure_Rate"], bins=20, kde=True)
plt.title("Distribution of Server Failure Rate")
plt.xlabel("Failure Rate (%)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="RAM_GB", y="Failure_Rate", hue="OS")
plt.title("RAM vs. Server Failure Rate")
plt.xlabel("RAM (GB)")
plt.ylabel("Failure Rate (%)")
plt.show()
```
